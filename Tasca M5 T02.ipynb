**INTRODUCCIÓ:**
-  El Train-Test Split s'utilitza per avaluar el rendiment d'un model. La seva funció principal és dividir el conjunt de dades disponible en dues parts: un conjunt d'entrenament (Train Set) i un conjunt de prova (Test Set).
-  'MEDV' serà la variable que utilitzaré en aquest exemple, com a target, per tal de predir-la.
-  La variable *x/features* seran totes les variables excepte "MEDV". I la variable *y/target* (objectiu) serà "MEDV".

```python
from sklearn.model_selection import train_test_split

X = data.drop('MEDV', axis=1)
y = data['MEDV']
# split into train test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

Després d'analitzar i comparar els conjunts graficament es podria dir que no hi ha diferencies significatives entre els dos conjunts.  El que si que es important de comentar es que en les variables : (NOX, RAD, TAX i PRATIO) hi trobem outliers en el test set. Posteriorment es tractaran amb l'aplicacio de transformacions...
Procedeixo a fer un test d'hipotesis per comprovar analiticament que no existeixen diferencies:

**Test de Shapiro-Wilk:**
- **Hipòtesi Nul·la (H0):** Les dades segueixen una distribució Normal
- **Hipòtesi Alternativa (H1):** Les dades NO segueixen una distribució Normal

```python
from scipy import stats
from scipy.stats import shapiro

for data in X_train.columns:
    shapiro_test = stats.shapiro(X_train[data])
    print( "variable:", data, "p-valor:", shapiro_test.pvalue)
    if shapiro_test.pvalue > 0.05:
        print("La mostra té una distribució normal.\n")
    else:
        print("La mostra No té una distribució normal.\n")
```

```python
for data in X_test.columns:
    shapiro_test = stats.shapiro(X_test[data])
    print( "variable:", data, "p-valor:", shapiro_test.pvalue)
    if shapiro_test.pvalue > 0.05:
        print("La mostra té una distribució normal.\n")
    else:
        print("La mostra No té una distribució normal.\n")
```

Tal i com indiquen els test de Shapiro concluim que ninguna variable segueix una distribució normal. Per comparar les dos mostres es pot utilizar el test de U de Mann-Whitney, que es una proba no parametrica utilitzada per comparar dues mostres independents i no requereix que hi hagi una distribució específica de les dades, per tant ideal en el nostre cas.

**Test de U de Mann-Whitney**

- **Hipòtesi Nul·la (H0):** La distribució de les dues mostres és la mateixa
- **Hipòtesi Alternativa (H1):** Hi ha diferències significatives entre les dues mostres

```python
from scipy.stats import mannwhitneyu

for variable in X_train.columns:
    statistic, p_value = mannwhitneyu(X_train[variable], X_test[variable])
    
    print(f"Variable: {variable}")
    print(f"Estadístic U: {statistic}")
    print(f"p-valor: {p_value}")
    
    if p_value < 0.05:
        print("Hi ha diferències significatives entre les dues mostres.\n")
    else:
        print("La distribució entre les dues mostres és la mateixa.\n")
```

Donat els resultats, podem dir que la funcio train_test_split ha funcionat prou bé, ja que no s'observen diferenecies significatives entre train i test menys en una variable que si que hi ha diferencies significatives entre les dues distribucions: 'PTRATIO'.

La transformació es fa en el training set perque representa que el test set no el conec. Pero abans faig una copia per tenir els valors de X_train originals (sense transformacions).

```python
X_train_transformat = X_train.copy() 
X_train_transformat.head(10)
```

```python
from sklearn import preprocessing 
X_train_transformat[["CRIM", "ZN", "RM", "DIS", "B", "LSTAT"]]=preprocessing.RobustScaler().fit_transform(X_train_transformat[["CRIM", "ZN", "RM", "DIS", "B", "LSTAT"]])
X_train_transformat.head(10)
```

Després d'elminar els outliers amb l'ús del robust scaler,transformaré les variables restants utilitzant el MinMaxScaler. 

```python
X_train_transformat[["INDUS","NOX","AGE","TAX","PTRATIO"]]=preprocessing.MinMaxScaler().fit_transform(X_train_transformat[["INDUS","NOX","AGE","TAX","PTRATIO"]])
X_train_transformat.head(10)
```

Convertiré la variable RAD en dummy perque es categorica:

```python
X_train['RAD'].value_counts()
```

```python
dummies = pd.get_dummies(X_train_transformat["RAD"], prefix="RAD", drop_first=True)
#elimino la columna RAD del X_train (conjunt entrenament)
X_train_transformat.drop("RAD", axis=1, inplace=True)

# Concateno las variables dummy al conjunt entrenament X train
X_train_transformat = pd.concat([X_train_transformat, dummies], axis=1)

X_train_transformat.head(10)
```
